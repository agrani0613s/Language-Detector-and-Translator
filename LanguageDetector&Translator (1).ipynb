```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("language.csv")
print(data.head())
```

```python
data["language"].value_counts()
```

```python
x=np.array(data["Text"])
y=np.array(data["language"])

print(x)
```

```python
print(y)
```

```python
cv=CountVectorizer()
X=cv.fit_transform(x)
x_train,x_test, y_train,y_test=train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
x_train
```

```python
print(x_train)
```

```python
y_train
```

```python
model_detect= MultinomialNB()
model_detect.fit(x_train, y_train)
```

```python
x_train_prediction = model_detect.predict(x_train)
training_data_accuracy= accuracy_score(x_train_prediction,y_train)
print("Accuracy on train data:",training_data_accuracy)
```

```python
x_test_prediction = model_detect.predict(x_test)
test_data_accuracy= accuracy_score(x_test_prediction,y_test)
print("Accuracy on test data:",test_data_accuracy)
```

```python
def detect_language(text):
    text_vectorized = cv.transform([text])  # Vectorize the input text
    predicted_language = model_detect.predict(text_vectorized)[0]  # Predict language
    return predicted_language
```

```python
print(detect_language("mi nombre es agrani"))
```

```python
from transformers import MarianMTModel, MarianTokenizer
import gradio as gr
```

```python

# Function to load the model and tokenizer for a language pair
def load_model(source_lang, target_lang):
    model_key = (source_lang, target_lang)
    if model_key in model_names:
        model_name = model_names[model_key]
        try:
            model_trans = MarianMTModel.from_pretrained(model_name)
            tokenizer = MarianTokenizer.from_pretrained(model_name)
            return model_trans, tokenizer
        except Exception as e:
            print(f"Error loading model {model_name}: {e}")
            return None, None
    else:
        print(f"Model for {source_lang} to {target_lang} not found.")
        return None, None

# Function to perform translation
def translate_text( text,source_lang, target_lang):
    model_trans, tokenizer = load_model(source_lang, target_lang)
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    translated = model_trans.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)
```

```python

# Available languages for the dropdown
languages = {
    'English': 'EN',
    'Spanish': 'ES',
    'French': 'FR',
    'Chinese': 'CN',
    'Vietnamese': 'VI',
    'Arabic': 'AR',
    'Korean': 'KO',
    'Japanese': 'JA',
    'German': 'DE',
    'Persian': 'FA',
    # Add more languages as needed
}

# Gradio function for translation
def gradio_translate(text, target_language):
    source_lang = detect_language(text) #Detecting source language
    source_lang_code = languages[source_lang]
    target_lang_code = languages[target_language]
    return source_lang, translate_text(  text, source_lang_code, target_lang_code)



```

```python
text="mi nombre es agrani"
source_lang=detect_language(text)
target_lang="French"
translated=gradio_translate(text,target_lang)
print(translated)
```

```python
# Create the Gradio interface
iface = gr.Interface(
    fn=gradio_translate,
    inputs=[
        gr.Textbox(label="Text to translate"),  # Text input
        # gr.Dropdown(list(languages.keys()), label="Source Language"),  # Source language dropdown
        gr.Dropdown(list(languages.keys()), label="Target Language")   # Target language dropdown
    ],
    outputs=[
        gr.Textbox(label="Detected Source Language", interactive=False),
        gr.Textbox(label="Translated Text"),
    ], # Translation output as text
    title="Multilingual Translation App",
    description="Translate text between multiple languages using Hugging Face models."
)
```

```python
# Launch the Gradio interface
iface.launch()
```
